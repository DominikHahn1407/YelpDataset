```python
import os

import numpy as np 
import pandas as pd
```

```python
chunk_size = 100000
```

```python
for dir_name, _, file_names in os.walk('data'):
    for file_name in file_names:
        print(os.path.join(dir_name, file_name))
```

```python
businesses = pd.read_json(os.path.join(dir_name, file_names[1]), lines=True, orient="columns", chunksize=chunk_size)
checkins = pd.read_json(os.path.join(dir_name, file_names[2]), lines=True, orient="columns", chunksize=chunk_size)
reviews = pd.read_json(os.path.join(dir_name, file_names[3]), lines=True, orient="columns", chunksize=chunk_size)
tips = pd.read_json(os.path.join(dir_name, file_names[4]), lines=True, orient="columns", chunksize=chunk_size)
users = pd.read_json(os.path.join(dir_name, file_names[5]), lines=True, orient="columns", chunksize=chunk_size)
```

```python
for business in businesses:
    business_chunck = business
    break

for checkin in checkins:
    checkin_chunk = checkin
    break

for review in reviews:
    review_chunk = review
    break

for tip in tips:
    tip_chunk = tip
    break

for user in users:
    user_chunk = user
    break

display(business_chunck.head(2))
display(checkin_chunk.head(2))
display(review_chunk.head(2))
display(tip_chunk.head(2))
display(user_chunk.head(2))
```
